```python
import folium
from folium import plugins
import pandas as pd
from ipywidgets import interact, widgets, Button, VBox, HBox, Layout
from IPython.display import display, clear_output

import numpy as np
import random
from folium.plugins import TagFilterButton

# Load the CSV file containing origin and destination GPS coordinates
df = pd.read_csv('ETT_v8_Origin_Destination.csv')

# Function to add lines and circles to the map
def add_lines_and_circles(filtered_df):
    m = folium.Map(control_scale=True, zoom_control=True)  # Create the map object without zoom_start
    
    for index, row in filtered_df.iterrows():
        try:
            origin = [float(row['GPS_Ori_Lat']), float(row['GPS_Ori_Lon'])]
            destination = [float(row['GPS_Des_Lat']), float(row['GPS_Des_Lon'])]
            origin_total = float(row['Origin_Total'])
            destination_total = float(row['Destination_Total'])
            des_dis = row['Des_Dis']
            
            # Create an AntPath for each pair of coordinates
            ant_path = plugins.AntPath(
                locations=[origin, destination], reverse=False, dash_array=[50, 40], color='blue', weight=1.5, delay=2000,
                tags=[des_dis]  # Add tag for filtering
            )
            ant_path.add_to(m)
            ant_path.add_child(folium.Popup(f"Des_Dis: {des_dis}"))
            ant_path.add_child(folium.Element(f'<div class="tag" data-tag="{des_dis}"></div>'))

            # Add circles at origin and destination with size based on total number
            origin_circle = folium.Circle(
                location=origin,
                radius=origin_total * 1,  # Adjust the multiplier to change circle size
                color='red',
                fill=True,
                fill_color='red',
                popup=f"Origin Total: {origin_total}, Des_Dis: {des_dis}",
                tags=[des_dis]  # Add tag for filtering
            )
            origin_circle.add_to(m)
            
            destination_circle = folium.Circle(
                location=destination,
                radius=destination_total * 1,  # Adjust the multiplier to change circle size
                color='yellow',
                fill=True,
                fill_color='yellow',
                popup=f"Destination Total: {destination_total}, Des_Dis: {des_dis}",
                tags=[des_dis]  # Add tag for filtering
            )
            destination_circle.add_to(m)
        except ValueError as e:
            print(f"Skipping row {index} due to error: {e}")

    # Fit the map to the bounds of the lines
    m.fit_bounds(m.get_bounds())
    
    # Set the zoom level after fitting the bounds
    m.options['zoom'] = 18

    return m

# Add TagFilterButton for 'Des_Dis' category
def add_tag_filter_button(m, df):
    tags = df['Des_Dis'].unique().tolist()
    TagFilterButton(data=tags, tags=tags, name='Des_Dis').add_to(m)

# Function to zoom to filtered items
def zoom_to_filtered_items(m, filtered_df):
    bounds = []
    for index, row in filtered_df.iterrows():
        origin = [float(row['GPS_Ori_Lat']), float(row['GPS_Ori_Lon'])]
        destination = [float(row['GPS_Des_Lat']), float(row['GPS_Des_Lon'])]
        bounds.extend([origin, destination])
    m.fit_bounds(bounds)


# Example usage
filtered_df = df  # Apply any filtering you need here
map_object = add_lines_and_circles(filtered_df)
add_tag_filter_button(map_object, df)
display(map_object)
```
